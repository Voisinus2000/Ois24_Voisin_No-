```r
# Charger les bibliothèques nécessaires
library(ggplot2)
library(reshape2)

# Utiliser le jeu de données mtcars
data(mtcars)

# Calculer les corrélations entre les variables
cor_matrix <- cor(mtcars)

# Convertir la matrice de corrélation en data frame long
melted_cor_matrix <- melt(cor_matrix)

# Créer la carte de chaleur
ggplot(data = melted_cor_matrix, aes(x = Var1, y = Var2, fill = value)) +
  geom_tile(color = "white") +
  scale_fill_gradient2(low = "blue", high = "red", mid = "white", 
                       midpoint = 0, limit = c(-1,1), space = "Lab", 
                       name="Correlation") +
  theme_minimal() + 
  theme(axis.text.x = element_text(angle = 45, vjust = 1, 
                                   size = 12, hjust = 1)) +
  coord_fixed() +
  labs(title = "Heatmap of Correlations in mtcars Dataset",
       x = "Variables",
       y = "Variables")
```
